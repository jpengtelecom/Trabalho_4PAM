```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
from scipy.special import erfc
from scipy import special
from scipy.fft import fft, fftfreq, ifft
from scipy.signal import firwin, lfilter
from scipy.io import wavfile
from scipy.io.wavfile import write
from scipy.signal import welch
from scipy import signal



# Parametros

N = 2000  # amostragem
p = 30
t = np.linspace(-p,p,N,endpoint=False)




# Portadora
data=(np.random.rand(10)>0.5)
data=(np.random.rand(10)>0.5)
portadora = np.repeat(data,200)

mensagem = np.random.randint(0, 4, N)
amplitude = [-3, -1, 1, 3]
sinal_mensagem = np.array([amplitude[i] for i in mensagem])



pam = sinal_mensagem*portadora #sinal PAM



# Plotagem
plt.figure(figsize=(18, 14))

plt.subplot(2, 1, 1)
plt.plot(t, sinal_mensagem)
plt.title('Sinal de Mensagem')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(t, pam)
plt.title('Sinal Modulado (4-PAM)')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
